```python
import pandas as pd
import yfinance as yf
from pathlib import Path
```

```python
base_path = Path("..")
universe_path = base_path / "data" / "external" / "universe.csv"
raw_path = base_path / "data" / "raw"
processed_path = base_path / "data" / "processed"

raw_path.mkdir(parents=True, exist_ok=True)
processed_path.mkdir(parents=True, exist_ok=True)

print(universe_path)
```

```python
universe = pd.read_csv(universe_path)
universe.head()
```

```python
tickers = universe["ticker"].tolist()
tickers
```

```python
start_date = "2018-01-01"
end_date = "2026-01-01"

data = yf.download(
    tickers=tickers,
    start=start_date,
    end=end_date,
    auto_adjust=False,
    group_by="ticker",
    progress=False
)

data.head()
```

```python
data.to_pickle(raw_path / "prices_raw.pkl")
print("Saved raw price data to:", raw_path / "prices_raw.pkl")
```

```python
data = pd.read_pickle(raw_path / "prices_raw.pkl")
data.head()
```

```python
data.columns
```

```python
data.columns.levels[0].tolist()
```

```python
records = []

for ticker in data.columns.levels[0]:
    temp = data[ticker].copy()

    # 如果整只股票全空，就跳过
    if temp.isna().all().all():
        print(f"Skipping empty ticker: {ticker}")
        continue

    temp["ticker"] = ticker
    temp = temp.reset_index()
    records.append(temp)

prices = pd.concat(records, ignore_index=True)
prices.head()
```

```python
prices.columns = [str(col).lower().replace(" ", "_") for col in prices.columns]
prices.head()
```

```python
print(prices.columns.tolist())
```

```python
prices = prices[["date", "ticker", "open", "high", "low", "close", "adj_close", "volume"]]
prices = prices.sort_values(["ticker", "date"]).reset_index(drop=True)
prices.head()
```

```python
print(prices.shape)
print(prices["ticker"].nunique())
print(prices["ticker"].unique())
```

```python
prices.isna().sum()
```

```python
coverage = prices.groupby("ticker").agg(
    start_date=("date", "min"),
    end_date=("date", "max"),
    rows=("date", "count")
).reset_index()

coverage
```

```python
coverage.to_csv(processed_path / "ticker_coverage.csv", index=False)
print("Saved ticker coverage report.")
```

```python
prices.to_csv(processed_path / "prices_clean.csv", index=False)
print("Saved cleaned data to:", processed_path / "prices_clean.csv")
```

```python
downloaded_tickers = sorted(prices["ticker"].unique().tolist())
downloaded_tickers
```

```python
universe_filtered = universe[universe["ticker"].isin(downloaded_tickers)].copy()
universe_filtered
```

```python
universe_filtered.to_csv(processed_path / "universe_filtered.csv", index=False)
print("Saved filtered universe.")
```
